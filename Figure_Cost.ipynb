```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
from matplotlib import style
```

```python
RC = (20505045, 17780785, 16824959)
RCC = (400791, 207050, 184942)
CC = (5936414, 5147713, 5031716)
CWDC = (55713, 3854, 2687)
TC = (26897964, 23139402, 23110589)
```

```python
xpos = np.arange(len(RC))
xpos
```

```python
style.use('fast')
plt.figure(figsize=(10,10))
barWidth = 0.2
plt.bar(xpos, RC, color='darkblue', width = barWidth, label='Rebar cost')
plt.bar(xpos+0.2, RCC, color='orange', width = barWidth, label='Rebar connection cost')
plt.bar(xpos+0.4, CC, color='darkgreen', width = barWidth, label='Carbon cost')
plt.bar(xpos+0.6, CWDC, color='yellow', width = barWidth, label='Waste disposal cost')
plt.bar(xpos+0.8, TC, color='red', width = barWidth, label='Total cost')
plt.xticks(xpos+0.3, ('Original', 'Previous', 'Proposed-Coupler'))
plt.xlabel('Methods')
plt.ylabel('Cost/Expense (USD)')
plt.title('Comparison in construction cost ')
plt.legend()
plt.savefig("figurecost.svg", bbox_inches="tight", transparent=True)
```
